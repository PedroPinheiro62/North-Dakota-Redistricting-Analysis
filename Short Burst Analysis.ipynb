# Short Burst Analysis

Read result files from Short Burst run and prints the max score found for each run and the minority's VAP population found for the partition with the best score in the run:

```python
import numpy as np
import pickle
import os

for file_name in os.listdir("data_sb/"):
    print(f"\n\nFile = {file_name}")
    if file_name.endswith(".npy"):
        res = np.load(f"data_sb/{file_name}")
        print(res)
        print(f"Max Number of Opp Districts: {np.max(res)}")

    else:
        with open(f"data_sb/{file_name}", "rb") as file:
            max_stats = pickle.load(file)
            print(max_stats["AMINVAP"])
            print(f"Max Stats = {max([max_stats["AMINVAP"][key] for key in max_stats["AMINVAP"].keys()])}")
```
