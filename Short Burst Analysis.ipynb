```python
import numpy as np
import pickle
import os

for file_name in os.listdir("data_sb/"):
    print(f"\n\nFile = {file_name}")
    if file_name.endswith(".npy"):
        res = np.load(f"data_sb/{file_name}")
        print(res)
        print(f"Max Score: {np.max(res)}")

    else:
        with open("data_sb/ND_dists47_AMINVAPopt_5.0%_40000_sbl15_scorenum_opportunity_dists_0.5.p", "rb") as file:
            max_stats = pickle.load(file)
            print(max_stats["AMINVAP"])
            print(f"Max Stats = {max([max_stats["AMINVAP"][key] for key in max_stats["AMINVAP"].keys()])}")
```
